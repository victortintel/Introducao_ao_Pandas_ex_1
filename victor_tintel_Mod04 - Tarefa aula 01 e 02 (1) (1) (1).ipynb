#### 1)  crie uma série do pandas a partir de uma lista com os dados abaixo:

Os dados originais podem ser vistos no site da NASA/GISS, e estão dispostos a cada década na tabela abaixo.

|ano|anomalia termica|
|:-:|:----:|
| 1900 | -0.08 |
| 1920 | -0.27 |
| 1940 | 0.12 |
| 1960 | -0.03 |
| 1980 | 0.26 |
| 2000 | 0.40 |
| 2020 | 1.02 |

Crie uma séries do Pandas a partir de uma lista com esses dados.

```python
import pandas as pd
import numpy as np

# Seu código aqui
anos = [1900, 1920, 1940, 1960, 1980, 2000, 2020]
anos_series_pandas = pd.Series(anos)
print(anos_series_pandas)
```

#### 5) Nomeie os índices das linhas com inteiros de 1 a 20, e as colunas com os nomes "x1", "x2", e "x3" respectivamente.

#### 6) No DataFrame do exercício 5, crie uma nova coluna como sendo a média das três colunas, e dê a ela o nome de "media" (não recomendo colocar acentos em nomes de variáveis).

```python
# seu código aqui
novos_dados = pd.DataFrame(np.random.normal(100, 10, (20,3)), columns=["x1", "x2", "x3"])
novos_dados["media"] = (novos_dados["x1"] + novos_dados["x2"] + novos_dados["x3"])/3
print(novos_dados)
```

#### 7) No DataFrame do exercício 6, crie uma nova coluna chamada "log_med", contendo o logaritmo natural da média calculada no exercício 6 <br>

```python
# seu código aqui
media_coluna = novos_dados["media"]
media_coluna
log_media = np.log(media_coluna)
log_media
```

```python
novos_dados["log_med"] = log_media
novos_dados
```
